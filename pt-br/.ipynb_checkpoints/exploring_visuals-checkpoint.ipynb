# Explorando com visualização
Use os arquivos `clean_08.csv` e `clean_18.csv`

```python
# Carregue os conjuntos de dados
# Importar as ferramentas e bibliotecas que iremos usar para criar gráficos

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
df_08.tail()
```

```python
#Compare as distribuições de classificação de emissão de gases do efeito estufa em 2008 e 2018.
df_08['greenhouse_gas_score'].plot(kind='hist', title='Greenhouse Gas 2008');
```

```python
df_18['greenhouse_gas_score'].plot(kind='hist', title='Greenhouse Gas 2018');
```

```python
# Como se alterou a distribuição de mpg combinado de 2008 para 2018?
df_08['cmb_mpg'].plot(kind='hist', title='Comb MPG 2008');
```

```python
df_18['cmb_mpg'].plot(kind='hist', title='Comb MPG 2018');
```

```python
# Descreva a correlação entre deslocamento (cilindrada) e mpg combinado.
df_08.plot(kind='scatter', x='cyl', y='cmb_mpg', title='Relação Cyl x Comb MPG 2008');
```

```python
df_18.plot(kind='scatter', x='cyl', y='cmb_mpg', title='Relação Cyl x Comb MPG 2018');
```

```python
# Descreva a correlação entre classificação de emissão de gases do efeito estufa e mpg combinado.
df_08.plot(kind='scatter', x='greenhouse_gas_score', y='cmb_mpg', title='Relação Emissão gás x Comb MPG 2008');
```

```python
df_18.plot(kind='scatter', x='greenhouse_gas_score', y='cmb_mpg', title='Relação Emissão gás x Comb MPG 2008');
```

```python
# Criar vetores de dataframe para 2008
list_08 = np.repeat("2008", df_08.shape[0])
```

```python
df_08["year"] = list_08
df_08.tail()
```

```python
# Criar vetores de dataframe para 2008
list_18 = np.repeat("2018", df_18.shape[0])
```

```python
df_18["year"] = list_18
df_18.head()
```

```python
final_df = df_08.append(df_18)
final_df.head()
```

```python
# Salve seus novos conjuntos de dados LIMPOS como novos arquivos!
final_df.to_csv('final_clean.csv', index=False)
final_df.tail()
```
